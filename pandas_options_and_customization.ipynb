```python
import numpy as np
import pandas as pd
```

# Display Maximum

```python
print(f'Maximum number of rows to display: {pd.get_option("display.max_rows")}')
print(f'Maximum number of columns to display: {pd.get_option("display.max_columns")}')
```

```python
df = pd.DataFrame(data=np.random.rand(100,100))
df
```

# Set Maximum Display

```python
pd.set_option("display.max_rows", 5)
pd.set_option("display.max_columns", 3)

print(f'Maximum number of rows to display: {pd.get_option("display.max_rows")}')
print(f'Maximum number of columns to display: {pd.get_option("display.max_columns")}')
```

```python
df = pd.DataFrame(data=np.random.rand(100,100))
df
```

# Reset Maximum Display

```python
pd.reset_option("display.max_rows")
pd.reset_option("display.max_columns")

print(f'Maximum number of rows to display: {pd.get_option("display.max_rows")}')
print(f'Maximum number of columns to display: {pd.get_option("display.max_columns")}')
```

```python
df = pd.DataFrame(data=np.random.rand(100,100))
df
```

# Describe Options

```python
pd.describe_option("display.max_rows")
pd.describe_option("display.max_columns")
```

# Precision

```python
pd.reset_option('display.precision')
with pd.option_context('display.precision', 10):
    print('Precision set:')
    print(pd.DataFrame(data=np.random.rand(2,2)))

print('\nPrecision not set:')
print(pd.DataFrame(data=np.random.rand(2,2)))
```
