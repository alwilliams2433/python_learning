```python
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
```

```python
xo = 0
xf = 1
yo = 1

x = np.linspace(xo, xf, 10)
y = yo * np.ones(x.shape)
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
y_noisy = y + np.random.uniform(low=-1, high=+1, size=y.shape)
```

```python
print(y_noisy)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, color='r', linestyle='-', linewidth=2, label='Analytical')
plt.plot(x, y_noisy, color='k', linestyle='-', linewidth=1, label='Noisy')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
def curve(x, lineParameter):
    return lineParameter * np.ones(x.shape)
```

```python
curve(x, 3)
```

```python
lineAdjusted, pcov = scipy.optimize.curve_fit(curve, x, y_noisy, p0=0.7)
print(lineAdjusted)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y_noisy, color='k', linestyle='-', linewidth=0.5, label='Noisy')
plt.plot(x, y, color='r', linestyle='-', linewidth=4, label='Analytical')
plt.plot(x, curve(x, lineAdjusted), color='b', linestyle='--', linewidth=4, label='Adjusted')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
