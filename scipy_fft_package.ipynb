```python
import numpy
import scipy.fftpack
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
final = 2*pi
x, dt = numpy.linspace(0, final, 50, retstep=True)
y = numpy.sin(x)
plt.plot(x, y)
plt.grid(True)
plt.xticks(numpy.arange(0, final+0.01, pi))
plt.show()

fft = scipy.fftpack.fft(y)
freqs = scipy.fftpack.fftfreq(len(y), d=dt)
plt.plot(freqs, numpy.real(fft))
plt.grid(True)
plt.show()
```
