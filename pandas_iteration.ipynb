```python
import pandas as pd
import numpy as np
```

```python
data = np.random.rand(10, 2)
index = np.random.randint(low=0, high=9, size=10)
df = pd.DataFrame(data=data, index=index, columns=['a', 'b'])
df
```

# Label Sorting

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```

# Value Sorting

```python
df.sort_values(by='a', kind='quicksort')
```

```python
df.sort_values(by=['a', 'b'])
```
