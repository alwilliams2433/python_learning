```python
import numpy
import scipy.fftpack
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
frequency = 1
numberOfRepetitions = 2
f = lambda x: numpy.sin(2*pi*frequency*x)
limits = (0., frequency*numberOfRepetitions)
numberOfPoints = 100*numberOfRepetitions # use even values
x, dx = numpy.linspace(limits[0], limits[1], numberOfPoints, retstep=True)
y = f(x)

fig = plt.figure(1, figsize=(12,6))
ax = fig.add_subplot(1,2,1)
ax.plot(x, y)
ax.set_title('sin function')
ax.set_xlabel('time')
ax.grid(True)

fft = scipy.fftpack.fft(y)
freqs = scipy.fftpack.fftfreq(numberOfPoints, d=dx)
ax = fig.add_subplot(1,2,2)
ax.plot(freqs[0:(numberOfPoints//2)], numpy.abs(fft[0:(numberOfPoints//2)]))
ax.set_xlim((0,2))
ax.set_title('sin function FFT')
ax.set_xlabel('frequency')
ax.grid(True)

plt.show()
```

```python
limits = (-pi, pi)
numberOfPoints = 500 # use even values
f = lambda x: 1 if x>=0 else -1
x, dx = numpy.linspace(limits[0], limits[1], numberOfPoints, retstep=True)
y = numpy.array([f(val) for val in x])

fig = plt.figure(2, figsize=(18,6))
ax = fig.add_subplot(1,3,1)
ax.plot(x, y)
ax.set_title('step function')
ax.set_xlabel('time')
ax.grid(True)

fft = scipy.fftpack.fft(y)
freqs = scipy.fftpack.fftfreq(numberOfPoints, d=dx)
ax = fig.add_subplot(1,3,2)
ax.plot(freqs[0:(numberOfPoints//2)], numpy.abs(fft[0:(numberOfPoints//2)]))
ax.set_xlim((0,1))
ax.set_xticks([(2*k-1)/(2*pi) for k in range(1, 10)])
ax.set_title('step function FFT')
ax.set_xlabel('frequency')
ax.grid(True)

ax = fig.add_subplot(1,3,3)
ax.plot(2*pi*freqs[0:(numberOfPoints//2)], numpy.abs(fft[0:(numberOfPoints//2)]))
ax.set_xlim((0,1))
ax.set_xticks([(2*k-1) for k in range(1, 10)])
ax.set_title('step function FFT')
ax.set_xlabel(r'$2 \pi f$')
ax.grid(True)

plt.show()
```

Note that in the above example, the amplitude is not important, but their relative values are.

It becomes clear in the next example where one of the signals is multiplied by $\dfrac{1}{2}$.

```python
frequencyA = 1
frequencyB = 2
numberOfRepetitions = 20
f = lambda x: numpy.sin(2*pi*frequencyA*x) + 0.5*numpy.sin(2*pi*frequencyB*x)
limits = (0., frequency*numberOfRepetitions)
numberOfPoints = 100*numberOfRepetitions # use even values
x, dx = numpy.linspace(limits[0], limits[1], numberOfPoints, retstep=True)
y = f(x)

fig = plt.figure(1, figsize=(12,6))
ax = fig.add_subplot(1,2,1)
ax.plot(x, y)
ax.set_title('compound sin function')
ax.set_xlabel('time')
ax.grid(True)

fft = scipy.fftpack.fft(y)
freqs = scipy.fftpack.fftfreq(numberOfPoints, d=dx)
ax = fig.add_subplot(1,2,2)
ax.plot(freqs[0:(numberOfPoints//2)], numpy.abs(fft[0:(numberOfPoints//2)]))
ax.set_xlim((0,4))
ax.set_title('compound sin function FFT')
ax.set_xlabel('frequency')
ax.grid(True)

plt.show()
```

```python
limits = (-pi, pi)
numberOfPoints = 500 # use even values
f = lambda x: 1 if x>=0 else -1
x, dx = numpy.linspace(limits[0], limits[1], numberOfPoints, retstep=True)
y = numpy.array([f(val) for val in x])
fft = scipy.fftpack.fft(y)
inverse_fft = scipy.fftpack.ifft(fft)
freqs = scipy.fftpack.fftfreq(numberOfPoints, d=dx)

fig = plt.figure(2, figsize=(18,6))
ax = fig.add_subplot(1,3,1)
ax.plot(x, y)
ax.set_title('step function')
ax.set_xlabel('time')
ax.grid(True)

ax = fig.add_subplot(1,3,2)
ax.plot(2*pi*freqs[0:(numberOfPoints//2)], numpy.abs(fft[0:(numberOfPoints//2)]))
ax.set_xlim((0,1))
ax.set_xticks([(2*k-1) for k in range(1, 10)])
ax.set_title('step function FFT')
ax.set_xlabel(r'$2 \pi f$')
ax.grid(True)

ax = fig.add_subplot(1,3,3)
ax.plot(x, numpy.real(inverse_fft))
ax.set_title('step function - inverse fft')
ax.set_xlabel('time')
ax.grid(True)

plt.show()
```

# Conclusion of how to use the FFT pack

1. Insert all your data. The more "repetitions" it has, the more precise will be the frequency domain transformation.
1. Plot your data as a function of the time just to be sure that it is something meaningful.
1. Apply the FFT: `y_frequency = scipy.fftpack.fft(y_time)`. That will be your y-axis data.
1. Get the FFT frequencies: `freqs = scipy.fftpack.fftfreq(numberOfPoints, d=dt)`.
Here `numberOfPoints=len(y_time)` and
`dt` is the time interval between two consecutive readings of the data aquisition system.
1. Now you have the "amplitudes" (relative amplitudes) and the frequencies.
To plot it, use:
 - `plot(freqs[0:(numberOfPoints//2)], numpy.abs(fft[0:(numberOfPoints//2)])`, for `numberOfPoints` even.
 - `plot(freqs[0:((numberOfPoints-1)//2)], numpy.abs(fft[0:((numberOfPoints-1)//2)])`, for `numberOfPoints` odd.
1. If necessary, it is possible to filter the data before plot it, just play with the FFT result.
