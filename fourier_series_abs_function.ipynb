```python
import numpy
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
x = numpy.linspace(-pi, pi, 100)
yf = numpy.abs(x)

fig = plt.figure(1, figsize=(8,8))

ax = fig.add_subplot(1,1,1)

ax.plot(x, yf, 'k-', linewidth=2, label=r'$f(x)=|x|$')

def FF(x, n):
    y = numpy.empty_like(x)
    for i, val in enumerate(x):
        y[i] = pi/2 - 4/pi*numpy.sum([[numpy.cos(k*val)/k**2] for k in numpy.arange(1,2*n,2)])
    return y
n = 1; ax.plot(x, FF(x,n), linestyle='-', color='red', label='n = {:d}'.format(n))
n = 2; ax.plot(x, FF(x,n), linestyle='--', color='red', label='n = {:d}'.format(n))
n = 3; ax.plot(x, FF(x,n), linestyle=':', color='red', label='n = {:d}'.format(n))

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.legend(loc='best')

plt.show()

def error(x, n):
    y = FF(x,n)
    return numpy.linalg.norm(FF(x,n) - numpy.abs(x))
for n in 2**numpy.arange(10):
    print('{:s} for n = {:3d} equals to {:.3e}'.format('Error norm log10', n, numpy.log10(error(x,n))))
```
