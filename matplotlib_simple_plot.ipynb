```python
import numpy
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
x = numpy.linspace(-pi, pi, 300, endpoint=True)
sine = numpy.sin(x)
cossine = numpy.cos(x)
```

```python
plt.plot(x,sine,x,cossine)
plt.show()
```

```python
plt.plot(x,sine,x,cossine)
plt.xticks([-pi, -pi/2, 0, pi/2, pi])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

```python
plt.plot(x,sine,x,cossine)
plt.xticks([-pi, -pi/2, 0, pi/2, pi],
           [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

```python
plt.plot(x,sine,x,cossine)
plt.xticks([-pi, -pi/2, 0, pi/2, pi],
           [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, -0.5, 0, 0.5, 1])
ax = plt.gca()
ax.spines["left"].set_color("none")
ax.spines["bottom"].set_color("none")
ax.spines["right"].set_position(("data",0))
ax.spines["top"].set_position(("data",0))
ax.xaxis.set_ticks_position("top")
ax.yaxis.set_ticks_position("right")
plt.show()
```

```python
plt.plot(x, sine, linestyle="-", color="r", linewidth=1, label="sine")
plt.plot(x, cossine, linestyle="-", color="b", linewidth=1, label="cossine")
plt.xticks([-pi, -pi/2, 0, pi/2, pi],
           [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, -0.5, 0, 0.5, 1])
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position(("data",0))
ax.spines["bottom"].set_position(("data",0))
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
plt.legend(loc="upper left", frameon=False)
plt.show()
```

```python
plt.plot(x, sine, linestyle="-", color="r", linewidth=1, label="sine")
plt.xticks([-pi, -pi/2, 0, pi/2, 2*pi/3, pi],
           [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$\pi/2$", r"$2\pi/3$", r"$\pi$"])
plt.yticks([-1, -0.5, 0, 0.5, numpy.sin(2*pi/3), 1])
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position(("data",0))
ax.spines["bottom"].set_position(("data",0))
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
plt.legend(loc="upper left", frameon=False)
val = 2*pi/3
plt.annotate(r"$sin\left(\frac{2\pi}{3}\right) = \frac{\sqrt{3}}{2}$",
            xy=(val,numpy.sin(val)), xycoords="data",
            xytext=(30, 10), textcoords="offset points",
            arrowprops=dict(arrowstyle="->"))
plt.plot(val, numpy.sin(val), marker=".", color="r", markersize=8)
plt.plot((val,val,0), (0,numpy.sin(val),numpy.sin(val)), linestyle="--", color="r", linewidth=1)
plt.show()
```

```python
plt.plot(x, sine, linestyle="-", color="r", linewidth=1, label="sine")
plt.xticks([-pi, -pi/2, pi/2, 2*pi/3, pi],
           [r"$-\pi$", r"$-\pi/2$", r"$\pi/2$", r"$2\pi/3$", r"$\pi$"])
plt.yticks([-1, -0.5, 0.5, numpy.sin(2*pi/3), 1],
           ["-1", "-0.5", "0.5", r"$\frac{\sqrt{3}}{2}$", "1"])
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position(("data",0))
ax.spines["bottom"].set_position(("data",0))
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(10)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.5 ))
plt.legend(loc="upper left", frameon=False)
val = 2*pi/3
plt.annotate(r"$sin\left(\frac{2\pi}{3}\right) = \frac{\sqrt{3}}{2}$",
            xy=(val,numpy.sin(val)), xycoords="data",
            xytext=(30, 10), textcoords="offset points",
            arrowprops=dict(arrowstyle="->"))
plt.plot(val, numpy.sin(val), marker=".", color="r", markersize=8)
plt.plot((val,val,0), (0,numpy.sin(val),numpy.sin(val)), linestyle="--", color="r", linewidth=1)
plt.show()
```
