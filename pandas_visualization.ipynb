```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(
    data = np.random.randn(10,4),
    index=pd.date_range('1/1/2000', periods=10, freq='M'),
    columns=list('ABCD'))
```

```python
df.plot()
```

```python
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.hist()
```
