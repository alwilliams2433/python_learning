```python
import numpy as np
import pandas as pd
```

# Names Example

```python
left = pd.DataFrame({
    'id'        : [1,      2,      3,       4,       5       ],
    'Name'      : ['Alex', 'Amy',  'Allen', 'Alice', 'Ayoung'],
    'subject_id': ['sub1', 'sub2', 'sub4',  'sub6',  'sub5'  ]
})
left
```

```python
right = pd.DataFrame({
    'id'        : [1,       2,       3,      4,       5      ],
    'Name'      : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'subject_id': ['sub2',  'sub4',  'sub3', 'sub6',  'sub5' ]
})
right
```

```python
pd.merge(left=left, right=right, on='id')
```

```python
pd.merge(left=left, right=right, how='outer')
```

# My example

```python
df_0 = pd.DataFrame(
    {
        'Goodness': np.random.choice(['Good', 'Bad'], size=(10,)),
        'Rank': np.random.choice(['A', 'B', 'C'], size=(10,))
    },
    index = np.arange(10)
)
df_0
```

```python
df_1 = pd.DataFrame(
    {
        'Goodness': np.random.choice(['Good', 'Bad'], size=(10,)),
        'Rank': np.random.choice(['A', 'B', 'C'], size=(10,))
    },
    index = 10 + np.arange(10)
)
df_1
```

```python
pd.concat([df_1, df_0])
```

```python
pd.concat([df_0, df_1], ignore_index=True)
```

```python
pd.concat([df_0, df_1], axis=1)
```

```python
pd.concat([df_0, df_1], axis=1, ignore_index=True)
```

```python
df_0.append(df_1)
```
