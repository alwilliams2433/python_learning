```python
import numpy
import matplotlib.pyplot as plt
import scipy.integrate
```

```python
f = lambda x: x**3 - 3*x + 4
limits = (0, 1)
numberOfPlots = 3
numberOfPoints = 1000
lambdas = 8**numpy.arange(1, numberOfPlots+1)
x = numpy.linspace(limits[0], limits[1], numberOfPoints)

fig = plt.figure(1, figsize=(6,6*numberOfPlots))
axs = fig.subplots(numberOfPlots, 1)

for i, ax in enumerate(axs):
    ax.plot(x, f(x), 'k-')
    ax.plot(x, -f(x), 'k--')
    ax.plot(x, f(x)*numpy.sin(lambdas[i]*x), 'r--')
    ax.spines['top'].set_color("None")
    ax.spines['right'].set_color("None")
    ax.spines['bottom'].set_position(("data",0))
    ax.grid(True)
    integralValue = scipy.integrate.simps(f(x)*numpy.sin(lambdas[i]*x), x)
    print('log10(|Integral|) = {:11.10e}'.format(numpy.log10(numpy.abs(integralValue))))

plt.show()
```
