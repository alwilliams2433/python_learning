```python
import numpy
from scipy.cluster.vq import kmeans,vq,whiten
import matplotlib.pyplot as plt
```

```python
data = numpy.random.rand(100,2)
numberOfClusters = 3
centroids,_ = kmeans(data, numberOfClusters)
# clusters,_ = vq(data,centroids)
plt.scatter(data[:,0], data[:,1], marker='o', color='b')
plt.scatter(centroids[:,0], centroids[:,1], marker='^', color='r')
plt.xlim([0,1])
plt.ylim([0,1])
plt.show()
```
