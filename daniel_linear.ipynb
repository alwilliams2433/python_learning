```python
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1_00)

def ramp(x, a, b):
    return a*x + b

y = ramp(x, 5, 10)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, color='r', linestyle='-', linewidth=2, label='Analytical')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
def f1(x):
    frequecy = 1
    return np.sin(2 * np.pi * frequecy * x)

def f2(x):
    frequecy = 10
    return np.sin(2 * np.pi * frequecy * x)

def f3(x):
    return f1(x) + f2(x)

plt.figure(figsize=(6,6))
plt.plot(x, f3(x), color='k', linestyle='-', label='Analytical')
plt.plot(x, f1(x), color='r', linestyle='-', label='f1')
#plt.plot(x, f2(x), color='b', linestyle='-', label='f2')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
y_noisy = y + np.random.uniform(low=-1, high=+1, size=y.shape)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, color='r', linestyle='-', linewidth=2, label='Analytical')
plt.plot(x, y_noisy, color='k', linestyle='-', linewidth=0.5, label='Noisy')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
listaA_B= [4,12]
lineAdjusted, pcov = scipy.optimize.curve_fit(ramp, x, y_noisy, p0=listaA_B)
print(lineAdjusted)
```

```python
a, b = lineAdjusted

plt.figure(figsize=(6,6))
plt.plot(x, y_noisy, color='k', linestyle='-', linewidth=0.5, label='Noisy')
plt.plot(x, y, color='r', linestyle='-', linewidth=4, label='Analytical')
plt.plot(x, ramp(x,a,b), color='b', linestyle='--', linewidth=4, label='Adjusted')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
