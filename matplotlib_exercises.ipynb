```python
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.fill_between(X, Y+1, np.ones_like(Y), color='blue', alpha=0.2)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.fill_between(X, Y-1, -1, where=((Y-1)>-1), color='blue', alpha=0.2)
plt.fill_between(X, Y-1, -1, where=((Y-1)<-1), color='red', alpha=0.2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = -(1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X, Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x, y+0.05, '%.2f' % y, ha='center', va='bottom')
for x,y in zip(X,Y2):
    plt.text(x, y-0.05, '{:.2f}'.format(y), ha='center', va='top')

plt.ylim(-1.25,+1.25)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.axes([0.025,0.025,0.95,0.95])
ax = plt.contourf(X, Y, f(X,Y), 7, alpha=.75, cmap='hot')
C = plt.contour(X, Y, f(X,Y), 7, colors='black', linewidths=.5)
plt.colorbar(ax)
plt.clabel(C, fontsize='smaller', inline=1)
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 20
Z = np.random.rand(n)
c = np.linspace(0,0.8,n)
colors = [(x,y,z) for x,y,z in zip(c,c,c)]
labels = ['{:.2f}'.format(x) for x in Z]
explode = np.zeros_like(Z); explode=explode+0.08; explode[-1]=0.3
plt.pie(Z, explode, shadow=True, labels=labels, colors=colors)
plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axes([0,0,1,1])

N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10*np.random.rand(N)
width = np.pi/4*np.random.rand(N)
ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.grid(True, which='major', linestyle=':')

for r,bar in zip(radii, bars):
    bar.set_facecolor( plt.cm.jet(r/10.))
    bar.set_alpha(0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)

plt.show()
```
