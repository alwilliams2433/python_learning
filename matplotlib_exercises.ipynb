```python
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.fill_between(X, Y+1, np.ones_like(Y), color='blue', alpha=0.2)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.fill_between(X, Y-1, -1, where=((Y-1)>-1), color='blue', alpha=0.2)
plt.fill_between(X, Y-1, -1, where=((Y-1)<-1), color='red', alpha=0.2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1024
X = np.random.rand(n)
Y = np.random.rand(n)
#angle = [np.angle(np.complex(x,y)) for x,y in zip(X-0.5,Y-0.5)]
angle = np.arctan2(Y-0.5,X-0.5)

plt.scatter(X, Y, c=angle, cmap='jet', edgecolors='k', alpha=0.4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = -(1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X, Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x, y+0.05, '%.2f' % y, ha='center', va='bottom')
for x,y in zip(X,Y2):
    plt.text(x, y-0.05, '{:.2f}'.format(y), ha='center', va='top')

plt.ylim(-1.25,+1.25)
plt.show()
```
