```python
import numpy
import scipy.stats
import matplotlib.pyplot as plt
```

```python
array = numpy.random.rand(5)
print(numpy.mean(array))
print(scipy.stats.mean(array))
# scipy is not a direct extension of numpy
```

```python
a = scipy.stats.alpha(1)
b = scipy.stats.bradford(1)
tz = scipy.stats.trapz(0, 1)
t = scipy.stats.t(1)
x = numpy.linspace(-2, 5, 100)
ya = a.pdf(x)
yb = b.pdf(x)
ytz = tz.pdf(x)
yt = t.pdf(x)
plt.plot(x, ya, 'b-', label='alpha')
plt.plot(x, yb, 'r-', label='bradford')
plt.plot(x, ytz, 'k-', label='trapezoidal')
plt.plot(x, yt, 'g-', label='t student')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
grades = numpy.array([
        10.0, 5.3, 8.4, 3.2
    ])
normal = scipy.stats.tmean(grades); print('{: <20s} = {:f}'.format('arithmetic mean', normal))
geometric = scipy.stats.gmean(grades); print('{: <20s} = {:f}'.format('geometric mean', geometric))
harmonic = scipy.stats.hmean(grades); print('{: <20s} = {:f}'.format('harmonic mean', harmonic))
```

```python
b = scipy.stats.binom(5, 0.4)
x = numpy.arange(-2, 6)
yb = b.pmf(x)
print(yb)
plt.plot(x, yb, 'bo', label='Binomial')
plt.legend(loc='best')
plt.show()
```
