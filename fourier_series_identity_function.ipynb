```python
import numpy
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
interval = (0.0, 1.0)
fx = lambda x: x
x = numpy.linspace(-interval[0], interval[1], 100)

fig = plt.figure(1, figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.plot(x, fx(x), 'k-', linewidth=2, label=r'$f(x)=|x|$')

def FF(x, n):
    y = numpy.empty_like(x)
    for i, val in enumerate(x):
        y[i] = 1/2 - 1/pi*numpy.sum([[numpy.sin(k*2*pi*val)/k] for k in numpy.arange(1,n+1)])
    return y
FFlabel = lambda n: 'n = {:d}'.format(n)
n = 1; ax.plot(x, FF(x,n), linestyle='-', color='red', label=FFlabel(n))
n = 3; ax.plot(x, FF(x,n), linestyle='--', color='red', label=FFlabel(n))
n = 5; ax.plot(x, FF(x,n), linestyle='-.', color='red', label=FFlabel(n))
n = 10; ax.plot(x, FF(x,n), linestyle=':', color='red', label=FFlabel(n))
n = 300; ax.plot(x, FF(x,n), linestyle='-', color='blue', label=FFlabel(n))

ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position(("data",0))
ax.spines["bottom"].set_position(("data",0))
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_xticks([0, 0.25, 0.5, 0.75, 1])
ax.set_xticklabels([r"$0$", r"$\dfrac{1}{4}$", r"$\dfrac{2}{4}$", r"$\dfrac{3}{4}$", r"$\dfrac{4}{4}=1$"])
ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.legend(loc="best", frameon=False)
ax.grid(True)

plt.show()

def error(x, n):
    y = FF(x,n)
    return numpy.linalg.norm(FF(x,n) - fx(x))
for n in 2**numpy.arange(10):
    print('{:s} for n = {:3d} equals to {:.3e}'.format('Error norm log10', n, numpy.log10(error(x,n))))
```
