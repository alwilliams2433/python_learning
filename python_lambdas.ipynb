```python
import matplotlib.pyplot as plt
import numpy
```

```python
a, b, c = 5, -3, 4
x = numpy.linspace(-1,1,50)
plt.plot(x,(lambda x: a*x**2 + b*x + c)(x))
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

f = lambda x,y : x*y
x = numpy.linspace(-1,1,50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = numpy.arange(-3.0, 3.0, 0.05)
X, Y = numpy.meshgrid(x, y)
zs = numpy.array([f(x,y) for x,y in zip(numpy.ravel(X), numpy.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)
plt.show()
```
