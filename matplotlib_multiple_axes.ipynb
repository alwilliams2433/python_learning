```python
import matplotlib.pyplot as plt
import numpy
print(plt.style.available)
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

fig = plt.figure(1, figsize=(8,8))

fig.suptitle('Three plots in the same figure'.title())

axes0 = fig.add_subplot(2,2,1)
axes0.bar(names, values)
axes0.set_title('bar graph'.title())
axes0.grid(True)

axes3 = fig.add_subplot(2,2,2)
axes3.semilogy(names, values)
axes3.set_title('log graph'.title())
axes3.grid(True)

axes1 = fig.add_subplot(2,2,3)
axes1.scatter(names, values)
axes1.set_title('scatter graph'.title())

axes2 = fig.add_subplot(2,2,4)
axes2.plot(names, values)
axes2.set_title('line graph'.title())

plt.show()
del fig
```

```python
x = numpy.linspace(0,10,50)
y = numpy.exp(x)

fig = plt.figure(1, figsize=(8,4))
fig.suptitle('log scale figure'.title())

ax0 = fig.add_subplot(1,2,1)
ax0.semilogy(x,y)
ax0.set_title('no grid'.title())

ax1 = fig.add_subplot(1,2,2)
ax1.semilogy(x,y)
ax1.yaxis.set_major_locator(plt.LogLocator(base=10))
ax1.grid(b=True, which='major', axis='y', alpha=1, color='g', linestyle='-')
ax1.grid(b=True, which='minor', axis='y', alpha=1, color='g', linestyle=':')
ax1.xaxis.set_major_locator(plt.MultipleLocator(1.0))
ax1.xaxis.set_minor_locator(plt.MultipleLocator(0.5))
ax1.grid(b=True, which='major', axis='x', alpha=1, color='m', linestyle='-')
ax1.grid(b=True, which='minor', axis='x', alpha=1, color='m', linestyle=':')
ax1.set_title('with grid'.title())

plt.show()
fig.savefig('opa.png')
del fig
```

```python
mu, sigma = 100, 15
x = mu + sigma * numpy.random.randn(100_000)
plt.hist(x, 100)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.grid(True)
plt.show()
```

```python
limits = 50
x = numpy.linspace(-limits,limits,100)
y1 = x
y2 = numpy.sin(x)

fig = plt.figure(1, figsize=(8,8))
ax = fig.subplots(2,2)

ax[0,0] = fig.add_subplot(2,2,1)
ax[0,0].plot(x,y1,'r', x,y2,'b')
ax[0,0].grid(True, which='both')
ax[0,0].set_xscale('linear')
ax[0,0].set_yscale('linear')

ax[0,1] = fig.add_subplot(2,2,2)
ax[0,1].plot(x,y1,'r', x,y2,'b')
ax[0,1].grid(True, which='both')
ax[0,1].set_xscale('linear')
ax[0,1].set_yscale('log')

ax[1,0] = fig.add_subplot(2,2,3)
ax[1,0].plot(x,y1,'r', x,y2,'b')
ax[1,0].grid(True, which='both')
ax[1,0].set_xscale('symlog')
ax[1,0].set_yscale('symlog', linthreshx=5.0)

ax[1,1] = fig.add_subplot(2,2,4)
ax[1,1].plot(x,y1,'r')
ax[1,1].plot(x,y2,'b')
ax[1,1].grid(True, which='both')
ax[1,1].set_xscale('linear')
ax[1,1].set_yscale('symlog', linthreshx=5.0)

del fig, x, y1, y2
```
