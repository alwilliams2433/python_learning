```python
import numpy
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
fig = plt.figure(1, figsize=(8,8))

ax = fig.add_subplot(1,1,1)

ax.plot([-pi, 0, 0, +pi], [-1, -1, +1, +1], 'k-', linewidth=2, label=r'$f(x)$')

for lastFourierTerm in numpy.arange(3, 8, 2):
    x = numpy.linspace(-pi, pi, 100)
    def fourierExpansion(x):
        y = numpy.empty_like(x)
        for i, val in enumerate(x):
            y[i] = (4/pi) * numpy.sum([[numpy.sin(k*val)/k] for k in numpy.arange(1,lastFourierTerm,2)])
        return y
    y = fourierExpansion(x)
    ax.plot(x, y, linestyle='-', color='gray', linewidth=1+0.3*lastFourierTerm, label='k={:d}'.format(lastFourierTerm-2))

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.legend(loc='best')

plt.show()
```
