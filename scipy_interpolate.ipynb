```python
import numpy
import scipy.interpolate
import matplotlib.pyplot as plt
pi = numpy.pi
```

```python
x = numpy.linspace(0,2*pi,12)
X = numpy.linspace(0,2*pi,50)
y = numpy.sin(x)
y1 = scipy.interpolate.interp1d(x, y, kind=1)
y2 = scipy.interpolate.interp1d(x, y, kind=2)
y3 = scipy.interpolate.interp1d(x, y, kind=3)
y5 = scipy.interpolate.interp1d(x, y, kind=5)
y9 = scipy.interpolate.interp1d(x, y, kind=9)
plt.plot(x, y, 'bo', label='data')
plt.plot(X, y1(X), 'g-', label='linear')
plt.plot(X, y2(X), 'r-', label='quadratic')
plt.plot(X, y3(X), 'c-', label='cubic')
plt.plot(X, y5(X), 'k-', label='5')
plt.plot(X, y5(X), 'y-', label='9')
plt.legend(loc='best')
print('log10(Error):')
print('1: ', numpy.log10(numpy.linalg.norm(numpy.sin(X) - y1(X))))
print('2: ', numpy.log10(numpy.linalg.norm(numpy.sin(X) - y2(X))))
print('3: ', numpy.log10(numpy.linalg.norm(numpy.sin(X) - y3(X))))
print('5: ', numpy.log10(numpy.linalg.norm(numpy.sin(X) - y5(X))))
print('9: ', numpy.log10(numpy.linalg.norm(numpy.sin(X) - y9(X))))
plt.show()
```
