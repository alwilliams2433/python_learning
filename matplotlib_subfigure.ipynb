```python
import matplotlib.pyplot as plt
import numpy
```

```python
plt.subplot(2,1,1)
x = numpy.linspace(-numpy.pi, numpy.pi)
y = numpy.sin(x)
plt.plot(x, y, 'g-')
a = plt.axes([0.3, 0.7, 0.1, 0.1], facecolor='gray')
plt.plot(x, y, 'g-')

plt.subplot(2,1,2)
x = numpy.linspace(-numpy.pi, numpy.pi)
y = -numpy.sin(x)
plt.plot(x, y, 'r-')
plt.show()

plt.close()
```

```python
fig = plt.figure(num=1, figsize=(8,8), dpi=50, facecolor='gray', edgecolor='g')
x = numpy.linspace(-numpy.pi, numpy.pi)
y = -numpy.sin(x)

axes = fig.add_subplot(2,2,1, facecolor='w')
axes.plot(x, y, 'g--')
axes.set_title('Axes 2, 2, 1')

axes = fig.add_subplot(2, 3, 3, facecolor='m')
axes.plot(x, y, 'k-')
axes.set_title('Trying magenta')

axes = fig.add_subplot(2,1,2, facecolor='k')
axes.plot(x, y, 'w-', label='sine')
axes.legend(loc='upper right')

fig.savefig('fig.png')
```
