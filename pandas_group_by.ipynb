```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'Kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
    'Rank': [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
    'Year': [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
    'Points':[876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby('Team')
```

```python
groups = df.groupby('Team').groups
df.loc[groups['Devils']]
```

```python
df.groupby('Team').get_group('Devils')
```

```python
df.groupby('Team')['Points'].aggregate([np.mean, np.sum, np.std])
```

```python
df.groupby('Team').filter(lambda x: np.mean(x['Year']) > 2015)
```
