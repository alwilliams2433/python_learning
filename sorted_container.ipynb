```python
sortedDict = sortedcontainers.SortedDict({'hello':'ola', 'hi':'oi', 'goodbye':'adeus', 'bye':'tchau'})
print(sortedDict)
print(sortedDict['hello'])
sortedDict['hello'] = 'SECRET'
print(sortedDict['hello'])
print(dict(zip([1],[2])))
```

```python
import numpy
size = 5
sortedDict = sortedcontainers.SortedDict(zip(numpy.random.rand(size), numpy.random.rand(size)))
print(sortedDict)
```

```python
sortedList = sortedcontainers.SortedList()
sortedList.add(1)
sortedList.add('hi')
# If we'd like to create a sorted container, we will
# never want to compare different objects!
```
